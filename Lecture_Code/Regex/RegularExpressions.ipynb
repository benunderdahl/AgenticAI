# 🔍 Regular Expressions in Python (`re` module)

## Importing
```python
import re
```

---

## Core Functions
| Function | Description |
|---|---|
| `re.match(pattern, string)` | Match **only at the beginning** of string |
| `re.search(pattern, string)` | Search **anywhere** in string, returns first match |
| `re.findall(pattern, string)` | Returns **list of all matches** |
| `re.finditer(pattern, string)` | Returns **iterator of match objects** |
| `re.sub(pattern, repl, string)` | **Replace** matches with a string |
| `re.split(pattern, string)` | **Split** string by pattern |
| `re.compile(pattern)` | **Compile** pattern for reuse |
| `re.fullmatch(pattern, string)` | Match **entire string** |

---

## Basic Usage
```python
import re

text = "Hello, my email is alice@example.com"

# search — find first match anywhere
match = re.search(r"\w+@\w+\.\w+", text)
if match:
    print(match.group())      # alice@example.com

# findall — get all matches as a list
re.findall(r"\d+", "I have 3 cats and 12 dogs")   # ['3', '12']

# sub — replace matches
re.sub(r"\s+", "-", "hello world foo")             # "hello-world-foo"

# split — split by pattern
re.split(r",\s*", "a, b,c,  d")                   # ['a', 'b', 'c', 'd']
```

---

## Match Object Methods
```python
match = re.search(r"(\d+)-(\w+)", "order 42-abc placed")

match.group()     # full match       → "42-abc"
match.group(1)    # first group      → "42"
match.group(2)    # second group     → "abc"
match.groups()    # all groups       → ("42", "abc")
match.start()     # start index      → 6
match.end()       # end index        → 12
match.span()      # (start, end)     → (6, 12)
```

## Pattern Syntax — Metacharacters
| Pattern | Meaning | Example Match |
|---|---|---|
| `.` | Any character except newline | `a.c` → `"abc"` |
| `^` | Start of string | `^Hello` → `"Hello world"` |
| `$` | End of string | `world$` → `"Hello world"` |
| `*` | 0 or more | `ab*` → `"a"`, `"ab"`, `"abbb"` |
| `+` | 1 or more | `ab+` → `"ab"`, `"abbb"` |
| `?` | 0 or 1 (optional) | `colou?r` → `"color"`, `"colour"` |
| `{n}` | Exactly n times | `\d{3}` → `"123"` |
| `{n,m}` | Between n and m times | `\d{2,4}` → `"12"`, `"1234"` |
| `[]` | Character class | `[aeiou]` → any vowel |
| `[^]` | Negated class | `[^aeiou]` → any non-vowel |
| `\|` | OR | `cat\|dog` → `"cat"` or `"dog"` |
| `()` | Capture group | `(\d+)` captures digits |
| `(?:)` | Non-capturing group | `(?:\d+)` groups without capturing |
| `\` | Escape special char | `\.` matches literal `.` |

---

## Character Classes (Shorthand)
| Pattern | Meaning | Equivalent |
|---|---|---|
| `\d` | Digit | `[0-9]` |
| `\D` | Non-digit | `[^0-9]` |
| `\w` | Word character | `[a-zA-Z0-9_]` |
| `\W` | Non-word character | `[^a-zA-Z0-9_]` |
| `\s` | Whitespace | `[ \t\n\r\f\v]` |
| `\S` | Non-whitespace | `[^ \t\n\r\f\v]` |
| `\b` | Word boundary | `\bcat\b` → `"cat"` not `"catch"` |
| `\B` | Non-word boundary | |

## Flags
```python
re.search(r"hello", "HELLO WORLD", re.IGNORECASE)   # case-insensitive
re.findall(r"^\d+", text, re.MULTILINE)              # ^ matches each line
re.search(r".+", text, re.DOTALL)                    # . matches newlines too
```

| Flag | Shorthand | Description |
|---|---|---|
| `re.IGNORECASE` | `re.I` | Case-insensitive matching |
| `re.MULTILINE` | `re.M` | `^` and `$` match each line |
| `re.DOTALL` | `re.S` | `.` matches newline too |
| `re.VERBOSE` | `re.X` | Allow comments in pattern |

## Compiled Patterns (for reuse)
```python
pattern = re.compile(r"\d{3}-\d{4}", re.IGNORECASE)

pattern.search("Call 555-1234 now")
pattern.findall("555-1234 or 888-5678")
pattern.sub("XXX-XXXX", "Call 555-1234")
```

## Groups & Named Groups
```python
# Numbered groups
match = re.search(r"(\w+)\s(\w+)", "John Doe")
match.group(1)    # "John"
match.group(2)    # "Doe"

# Named groups
match = re.search(r"(?P<first>\w+)\s(?P<last>\w+)", "John Doe")
match.group("first")    # "John"
match.group("last")     # "Doe"
match.groupdict()       # {"first": "John", "last": "Doe"}
```

## Common Real-World Patterns
```python
# Email
re.search(r"[\w.-]+@[\w.-]+\.\w{2,}", text)

# Phone number
re.search(r"\(?\d{3}\)?[-.\s]\d{3}[-.\s]\d{4}", text)

# URL
re.search(r"https?://[\w./%-]+", text)

# IP Address
re.search(r"\b\d{1,3}(\.\d{1,3}){3}\b", text)

# Date (YYYY-MM-DD)
re.search(r"\d{4}-\d{2}-\d{2}", text)

# Zip Code
re.search(r"\b\d{5}(-\d{4})?\b", text)
```

---

## Greedy vs Non-Greedy
```python
text = "<h1>Title</h1>"

re.search(r"<.+>", text).group()    # "<h1>Title</h1>"  (greedy)
re.search(r"<.+?>", text).group()   # "<h1>"            (non-greedy)
```
> Add `?` after a quantifier (`*?`, `+?`, `??`) to make it non-greedy.

## Tips
- Always use **raw strings** `r"pattern"` to avoid issues with backslashes
- Use `re.compile()` when using the same pattern multiple times
- Test patterns interactively at **regex101.com**
- Use `re.VERBOSE` with comments for complex patterns:
```python
pattern = re.compile(r"""
    (\d{3})   # area code
    [-.\s]    # separator
    (\d{3})   # prefix
    [-.\s]    # separator
    (\d{4})   # line number
""", re.VERBOSE)
```

````python
apple_search = re.search(r'[abc]', 'apple') #searching for any of the characters in the brackets for given input
# there is a match of a in this case

apple_findall = re.findall(r'[abc]', 'apple sauce') # we will get all matches for findall there are two a matches and one c
print(apple_search)
print(apple_findall)
# quick not characters in brackets are case sensitive

# comparing match , search and findall
txt = "Welcome to SkillStorm"
match1 = re.match(r'[lSk]', txt) # look at beginning of string is starts with l S or k there is a match
search1 = re.search(r'[lSk]', txt) # searches whole string and if any chars match return the first match
findall = re.findall(r'[lSk]', txt) # return all the matchs in teh string that match any chars in brackets
print(match1) # none
print(search1) # span of first match starting index 2 inclusive ending index 3 exclusive
print(findall) # returns all instances that match perhaps for counting number of matches
````

````python
"""
Meta Characters 

[] - a set of chars
. - any char
^ - starts with || NOT
* - Zero or more
+ - One or more
? - Zero or one 
() - allow us to group chars
| - either or

{char} - match exactly n occurences
"""
txt = "foofoofoofoofoofoofoo"
search2 = re.search("(foo){3}", txt)
print(search2) # we have a match because we have 3 groupings of foo just like we are looking for


txt = "Welcome to SkillStorm"
match2 = re.match(r'[\w]+', txt) # match any char 1 to unlimited because of + 
print(match2) # returns span of match of any chars that are alphanumeric

findall1 = re.findall(r'[\w]+', txt)
print(findall1) # outputs welcome to skillstorm in 3 items in a list because space is non alphanumeric so 
# the findall will return all matches similar to split() function in python


````

````python
# look behind / look ahead

'''
    used when we want to look for a pattern before or after our current match
    probably used for email validation 
    syntax : (?<=abc) -- match string that is preceded by abc
    syntax : (?=abc) -- match the string that is followed by abc
'''

txt = "Hello : Welcome to Skillstorm"
matches = re.findall(r"(?<=:).*", txt)
print(matches) # returns everything that is preceded by : includes whitespace before Welcome

matches = re.findall(r".+(?=:)", txt) 
print(matches) # outputs Hello with whitespace because we are loking for one or more occurences because of +

# use escape chars or r to tell python to treat string as a raw string
````

````python
# using split and sub
# split will return a list splitting by pattern 
# sub will replace any number of matched patterns in a string

dummy_text = """
Lorem ipsum dolor sit amet consectetur adipiscing elit. 
Quisque faucibus ex sapien vitae pellentesque sem placerat. 
In id cursus mi pretium tellus duis convallis. 
Tempus leo eu aenean sed diam urna tempor.
 Pulvinar vivamus fringilla lacus nec metus bibendum egestas. 
Iaculis massa nisl malesuada lacinia integer nunc posuere.
Ut hendrerit semper vel class aptent taciti sociosqu. 
Ad litora torquent per conubia nostra inceptos himenaeos.
"""

split_at_period = re.split("\\.+", dummy_text) # splits the string at every period,
# since . is a regex char matching pattern escape it with \\
print(split_at_period)

# sub will continue throughtout the whole string
sub_text = re.sub(" ", "&&", dummy_text) # replace every white space with &&
print(sub_text)
````
