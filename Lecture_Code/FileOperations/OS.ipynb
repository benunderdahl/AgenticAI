# Interacting with our OS directly using python

```python
# to work with OS we import os , part of python standard library
import os

cwd = os.getcwd() # os method to get current working directory
print(cwd)

# changing directories 
os.chdir("../") # go one folder up

cwd = os.getcwd()
print(cwd) # displays path one folder up from previous
```

```python
# Likewise creating new directories can be done using os module

os.mkdir("./" + "/demo") # adds a demo folder into the path that was specified above
# since the path is mutable it will stay the same

# if not using relative paths would probably be better to store these paths in variables
os.mkdir("./demo1")
os.mkdir("./demo2")
# will get errors if the directories already exist
```

```python
# can also join paths for modularity
# ex having a base path and multiple routes to new paths by joining

base_path = "./" # relative path

demo_path = os.path.join(base_path + "demo")
demo1_path = os.path.join(base_path + "demo1") # can insert using string or variable
demo2_path = os.path.join(base_path + "demo2")
```

```python
# getting all directories in a list
# specifying the root directory

root = "/" # root directory on computer ie linux command cd /
# we can list all directories using listdir

os.listdir(root) # lists all directories from the root or whatever path specified
# like a tree and will go down recursively
```

```python
# if we can create and modify we can also delete
# deleting the folders that were made above 
# already have paths saved in variable from join function
# only empty directories can be removed using method below
os.rmdir(demo_path)
os.rmdir(demo1_path)
os.rmdir(demo2_path)

# likewise if it doesnt exist there will be an error raised
# nested directories can be removed if they are also empty
```

```python
# removing files

# creating some dummy file that we will use to delete using os module
#os.mkdir(base_path + "deleteme")
with open(base_path + "deleteme/delete.txt", "w") as f:
    f.write("dummy data") # since it doesnt exist python will create this file

# removing this file using remove command
os.remove(base_path + "deleteme/delete.txt") # remove the file using path
# folder is now empty we can delete folder using rmdir
os.rmdir(base_path + "deleteme")
```

```python
# reading multiple files and directories using walk

for root, dirs, files in os.walk(os.getcwd()):
    print(root) # gives current root
    print(dirs) # directories that exist in the root specified
    print(files) # gives files inside directories

    # dir and files will be output into lists
```
