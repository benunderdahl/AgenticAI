# Error Handling

```python
# in python done with try/except blocks and finally will always execute
# unless sys exit is invoked

try: 
    print("Try block")
except: # catches every exception has a broad scope, can narrow scope
    print("There was an error")
```

```python
# useful for file reading to catch errors that can
# happen such as reading from a file that doesnt exist

try:
    f = open("./nonexistent_file.txt")
except IOError:
    print("There is no such file")
finally:
    print("Finally will still execute after catching exception")
```

```python
# catching narrow types of exceptions such as division by zero

x = 10
y = 0

try:
    z = x / y
except ArithmeticError: # looking for a specific exception rather than generalizing
    print("Divisin by zero not allowed")
finally:
    print("Im still going to execute")
```

```python
# just like in java we can thow exceptions
# in python the syntax is raise

x = "String"
try:
    if not type(x) is int:
        raise TypeError("Should be an int")
except TypeError as e:
    print(e)
```
