# File Operations

# 📁 File Operations in Python

## Opening a File
```python
open(filepath, mode, encoding)
```

| Mode | Description |
|---|---|
| `"r"` | Read (default) — error if file doesn't exist |
| `"w"` | Write — creates file, **overwrites** if exists |
| `"a"` | Append — creates file, adds to end if exists |
| `"x"` | Create — error if file already exists |
| `"rb"` | Read binary (images, pdfs, etc.) |
| `"wb"` | Write binary |

---

## Always Use `with` (Context Manager)
Automatically closes the file even if an error occurs.
```python
with open("file.txt", "r") as f:
    content = f.read()
```

---

## Reading Files
```python
# Read entire file as one string
with open("file.txt", "r") as f:
    content = f.read()

# Read line by line into a list
with open("file.txt", "r") as f:
    lines = f.readlines()        # ["line1\n", "line2\n"]

# Read one line at a time
with open("file.txt", "r") as f:
    line = f.readline()

# Iterate line by line (memory efficient for large files)
with open("file.txt", "r") as f:
    for line in f:
        print(line.strip())
```

---

## Writing Files
```python
# Write (overwrites existing content)
with open("file.txt", "w") as f:
    f.write("Hello, World!\n")

# Write multiple lines
with open("file.txt", "w") as f:
    lines = ["line1\n", "line2\n", "line3\n"]
    f.writelines(lines)

# Append to existing file
with open("file.txt", "a") as f:
    f.write("New line at the end\n")
```

## Working with Paths (pathlib — recommended)
```python
from pathlib import Path

# Common paths
Path(".")                          # current directory
Path.home()                        # /Users/username
Path.cwd()                         # current working directory

# Building paths
base = Path.home() / "Documents" / "project"
file = base / "data.txt"

# Path properties
file.name           # "data.txt"
file.stem           # "data"
file.suffix         # ".txt"
file.parent         # /Users/username/Documents/project

# Checks
file.exists()       # True/False
file.is_file()      # True/False
file.is_dir()       # True/False

# Read & write directly with pathlib
file.write_text("Hello!")
content = file.read_text()
```

## File & Directory Operations (os & shutil)
```python
import os
import shutil
from pathlib import Path

# Create a directory
os.makedirs("new_folder", exist_ok=True)       # exist_ok avoids error if exists
Path("new_folder").mkdir(parents=True, exist_ok=True)  # pathlib version

# Rename a file
os.rename("old.txt", "new.txt")

# Delete a file
os.remove("file.txt")

# Delete a folder
os.rmdir("empty_folder")                       # only works if folder is empty
shutil.rmtree("folder_with_contents")          # deletes folder and everything in it

# Copy a file
shutil.copy("source.txt", "destination.txt")

# Move a file
shutil.move("source.txt", "new/location/source.txt")

# List files in a directory
os.listdir(".")                                # returns list of names
list(Path(".").iterdir())                      # pathlib version
```

## Checking & Searching Files
```python
import os
from pathlib import Path

# Check if file exists
os.path.exists("file.txt")
Path("file.txt").exists()

# Get file size (bytes)
os.path.getsize("file.txt")
Path("file.txt").stat().st_size

# List only .txt files in a folder
list(Path(".").glob("*.txt"))

# Search recursively through subfolders
list(Path(".").rglob("*.txt"))
```

## Working with File Paths (os.path)
```python
import os

os.path.join("folder", "file.txt")        # "folder/file.txt"
os.path.basename("/Users/john/file.txt")  # "file.txt"

os.path.splitext("file.txt")             # ("file", ".txt")
```

## Reading CSV Files
```python
import csv

# Read
with open("data.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)                        # each row is a list

# Read as dictionary
with open("data.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)                        # each row is a dict

# Write
with open("data.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["name", "age"])      # header
    writer.writerow(["Alice", 30])
```

## Reading JSON Files
```python
import json

# Read JSON file
with open("data.json", "r") as f:
    data = json.load(f)

# Write JSON file
with open("data.json", "w") as f:
    json.dump(data, f, indent=4)

# String ↔ JSON
json_string = json.dumps(data)            # dict → string
data = json.loads(json_string)            # string → dict
```

## Exception Handling with Files
```python
try:
    with open("file.txt", "r") as f:
        content = f.read()
except FileNotFoundError:
    print("File does not exist")
except PermissionError:
    print("No permission to access file")
except IsADirectoryError:
    print("Expected a file, got a directory")
except Exception as e:
    print(f"Unexpected error: {e}")
```

## Tips
- Always use `with open()` — it handles closing the file automatically
- Use `pathlib` over `os.path` for cleaner, more readable path handling
- Use `encoding="utf-8"` explicitly when reading text files to avoid encoding issues
- Use `newline=""` when writing CSV files on Windows to avoid extra blank lines
- For large files, iterate line by line instead of loading the whole file into memory
