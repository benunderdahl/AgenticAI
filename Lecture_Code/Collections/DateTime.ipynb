````python
# Datetime module
# Creates date time objects
# has limited support for timezones

import datetime 

# Formatting : year, month, day, hour, minute, seconds, microsecond

print(datetime.datetime.now())

print(datetime.date.today()) # Output : YYYY-MM-DD

print(type(datetime.date.today())) # of class datetime.date
````

````python
            # Formatting Datetime objects
# use cases for using formatting
# America: month- day- year
# Europe : day - month- year

now = datetime.datetime.now()

print(now.strftime('%a')) #returns the day of datetime object
print(now.strftime('%d %m %y')) # returns day, month and year YY

# formatting for possible eu db
print(now.strftime('%d-%m-%Y')) # returns day, month, year YYYY

# formatting for possible db in america somewhere
print(now.strftime('%m-%d-%Y')) # returns month, day year YYYY



````

````python
# Parsing string into a datetime object

date = '2026-04-01'
date_obj = datetime.datetime.strptime(date, '%Y-%m-%d') #parses string into datetime object also matches format of string
print(date_obj) # prints parsed string in object with time of 00:00:00
````

# 📅 Python Datetime API

## Importing
```python
from datetime import datetime, date, time, timedelta
import datetime as dt
```

## Creating Dates & Times
```python
today = date.today()                        # current date
now = datetime.now()                        # current date + time
d = date(2024, 1, 15)                       # specific date
t = time(14, 30, 0)                         # specific time (2:30 PM)
dt = datetime(2024, 1, 15, 14, 30, 0)      # specific datetime
```

## Accessing Components
```python
now.year        # 2024
now.month       # 1
now.day         # 15
now.hour        # 14
now.minute      # 30
now.second      # 0
now.weekday()   # 0=Monday, 6=Sunday
```

## Formatting Dates (strftime)
```python
now.strftime("%Y-%m-%d")           # "2024-01-15"
now.strftime("%d/%m/%Y")           # "15/01/2024"
now.strftime("%B %d, %Y")         # "January 15, 2024"
now.strftime("%I:%M %p")          # "02:30 PM"
now.strftime("%Y-%m-%d %H:%M:%S") # "2024-01-15 14:30:00"
```

## Parsing Strings into Dates (strptime)
```python
datetime.strptime("2024-01-15", "%Y-%m-%d")
datetime.strptime("15/01/2024", "%d/%m/%Y")
datetime.strptime("January 15, 2024", "%B %d, %Y")
```

## Common Format Codes
| Code | Meaning | Example |
|---|---|---|
| `%Y` | 4-digit year | `2024` |
| `%m` | Month (zero-padded) | `01` |
| `%d` | Day (zero-padded) | `15` |
| `%H` | Hour 24hr | `14` |
| `%I` | Hour 12hr | `02` |
| `%M` | Minutes | `30` |
| `%S` | Seconds | `00` |
| `%p` | AM/PM | `PM` |
| `%A` | Weekday name | `Monday` |
| `%B` | Month name | `January` |

## timedelta — Date Arithmetic
```python
from datetime import timedelta

today = date.today()
tomorrow = today + timedelta(days=1)
last_week = today - timedelta(weeks=1)
in_2_hours = datetime.now() + timedelta(hours=2)

# Difference between two dates
d1 = date(2024, 1, 1)
d2 = date(2024, 6, 1)
diff = d2 - d1
diff.days    # 152
```

## Comparing Dates
```python
d1 = date(2024, 1, 1)
d2 = date(2024, 6, 1)

d1 < d2    # True
d1 == d2   # False
d1 > d2    # False
```

## Timestamps
```python
# datetime → timestamp
now = datetime.now()
ts = now.timestamp()      # float (Unix timestamp)

# timestamp → datetime
datetime.fromtimestamp(ts)
```

## Timezone Awareness (zoneinfo — Python 3.9+)
```python
from zoneinfo import ZoneInfo

utc_now = datetime.now(tz=ZoneInfo("UTC"))
ny_now = datetime.now(tz=ZoneInfo("America/New_York"))
london = datetime.now(tz=ZoneInfo("Europe/London"))
```
