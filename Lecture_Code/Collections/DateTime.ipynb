```python
# Datetime module
# Creates date time objects
# has limited support for timezones

import datetime 

# Formatting : year, month, day, hour, minute, seconds, microsecond

print(datetime.datetime.now())

print(datetime.date.today()) # Output : YYYY-MM-DD

print(type(datetime.date.today())) # of class datetime.date
```

```python
            # Formatting Datetime objects
# use cases for using formatting
# America: month- day- year
# Europe : day - month- year

now = datetime.datetime.now()

print(now.strftime('%a')) #returns the day of datetime object
print(now.strftime('%d %m %y')) # returns day, month and year YY

# formatting for possible eu db
print(now.strftime('%d-%m-%Y')) # returns day, month, year YYYY

# formatting for possible db in america somewhere
print(now.strftime('%m-%d-%Y')) # returns month, day year YYYY



```

```python
# Parsing string into a datetime object

date = '2026-04-01'
date_obj = datetime.datetime.strptime(date, '%Y-%m-%d') #parses string into datetime object also matches format of string
print(date_obj) # prints parsed string in object with time of 00:00:00
```
