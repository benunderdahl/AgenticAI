# 🗂️ Python Collections

## Lists
Ordered, mutable, allows duplicates.
```python
lst = [1, 2, 3, "hello", True]
lst = list((1, 2, 3))
```

### Indexing & Slicing
```python
lst[0]       # first element
lst[-1]      # last element
lst[1:4]     # slice index 1 to 3
lst[::-1]    # reverse list
```

### Common List Methods
| Method | Description | Example |
|---|---|---|
| `lst.append(x)` | Add to end | `[1,2].append(3)` → `[1,2,3]` |
| `lst.insert(i, x)` | Insert at index | `[1,3].insert(1,2)` → `[1,2,3]` |
| `lst.extend(iterable)` | Add multiple items | `[1].extend([2,3])` → `[1,2,3]` |
| `lst.remove(x)` | Remove first occurrence | `[1,2,2].remove(2)` → `[1,2]` |
| `lst.pop(i)` | Remove & return at index | `[1,2,3].pop(0)` → `1` |
| `lst.index(x)` | Find index of value | `[1,2,3].index(2)` → `1` |
| `lst.count(x)` | Count occurrences | `[1,2,2].count(2)` → `2` |
| `lst.sort()` | Sort in place | `[3,1,2].sort()` → `[1,2,3]` |
| `lst.reverse()` | Reverse in place | `[1,2,3].reverse()` → `[3,2,1]` |
| `lst.copy()` | Shallow copy | `[1,2].copy()` → `[1,2]` |
| `lst.clear()` | Remove all items | `[1,2].clear()` → `[]` |

### Useful Built-ins with Lists
```python
len(lst)           # number of elements
sum(lst)           # sum of elements
min(lst)           # smallest element
max(lst)           # largest element
sorted(lst)        # returns new sorted list
reversed(lst)      # returns reversed iterator
enumerate(lst)     # index + value pairs
zip(lst1, lst2)    # pair up two lists
```

### List Comprehensions
```python
squares = [x**2 for x in range(10)]
evens   = [x for x in range(10) if x % 2 == 0]
flat    = [x for row in matrix for x in row]
```


# Lists

````python

# declaring and instantiating a list of strings
str_list = ["George", "Ringo", "John", "Paul"]

# splitting up a list can be done with {start index}:{end index}
# by default if no value is specified after : then it will go to the end 
# of the list

print(str_list[1:2]) #starts at index 1 inclusive and ends at 2 exclusive returns ringo
print(str_list[1:]) # since no end index specified it will return ringo, john, paul

# lists are full of useful functions such as reverse and shortcuts
# if you use [::-1] it will return the same list in reverse
print(str_list[::-1])


# reversing a list can also be done using reversed()
#print(list(reversed(str_list))) #reversed() returns an iterator and list() converts the iterator back to a list

# creating sublists by slicing lists
sublist = str_list[1:]
print(sublist) # contains everything from index 1 to the end

# apeending to a list 
# because lists are mutable the actual list object will change
str_list.append("A new Beatle")
print(str_list) # prints all beatles and a new beatle

# by default pop method will remove the last value in the list and return it
removed_beatle = str_list.pop()
print(removed_beatle) # returns 'A new beatle'

# we can insert and remove from a list specifying index
str_list.remove("John") # remove John from the list
str_list.insert(0, "John") # insert the valu "John " at index 0
print(str_list) # will output John, George,ringo, paul

# we can find values with iterators and the in keyword
if ("John" in str_list):
    print(True)

# lists can be merged using + operator
list1 = [1,2,3,4]
list2 = [5,6,7,8]
print(list1 + list2) # outputs 1-8 in essence list is appended 

# enumerating a list with enumerate()
for i, element in enumerate(str_list):
    print(i, element) # i stores index element is actual value

# wrting lists based off boolean expressions using list comprehension
beatles = ["John", "Paul", "Ringo", "George"]
new_beastles = [name for name in beatles if name != "John"] #returns all beatles without John
print(new_beastles)

````

## Sets
Unordered, **no duplicates**, mutable.
```python
s = {1, 2, 3}
s = set([1, 2, 2, 3])    # → {1, 2, 3}
s = set()                 # empty set (NOT {})
```

### Common Set Methods
| Method | Description | Example |
|---|---|---|
| `s.add(x)` | Add element | `{1}.add(2)` → `{1,2}` |
| `s.remove(x)` | Remove (raises error if missing) | `{1,2}.remove(1)` → `{2}` |
| `s.discard(x)` | Remove (no error if missing) | `{1,2}.discard(9)` → `{1,2}` |
| `s.pop()` | Remove & return arbitrary element | |
| `s.clear()` | Remove all elements | |
| `s.copy()` | Shallow copy | |

### Set Operations
```python
a = {1, 2, 3}
b = {3, 4, 5}

a | b             # union          → {1,2,3,4,5}
a & b             # intersection   → {3}
a - b             # difference     → {1,2}
a ^ b             # symmetric diff → {1,2,4,5}

a.issubset(b)     # is a ⊆ b?
a.issuperset(b)   # is a ⊇ b?
a.isdisjoint(b)   # no elements in common?
```



````python
# they are unordered

# string will store unique values for a given collection
my_set = {1,2,3,4,5,6, 1, 2}
get_unique_in_string = "this and this and this and this"

print(my_set) # will print 1 - 6 not duplicates

print(set(get_unique_in_string.split())) # splitting each word and using set to get unique values
# when run it wil print {'this', 'and'}
````

## Tuples
Ordered, **immutable**, allows duplicates.
```python
t = (1, 2, 3)
t = tuple([1, 2, 3])
t = (42,)          # single element tuple (comma required)
```

### Common Tuple Operations
| Operation | Description | Example |
|---|---|---|
| `t[0]` | Indexing | `(1,2,3)[0]` → `1` |
| `t[1:3]` | Slicing | `(1,2,3)[1:3]` → `(2,3)` |
| `t.count(x)` | Count occurrences | `(1,2,2).count(2)` → `2` |
| `t.index(x)` | Find index | `(1,2,3).index(3)` → `2` |
| `t1 + t2` | Concatenate | `(1,2)+(3,4)` → `(1,2,3,4)` |
| `t * n` | Repeat | `(1,2)*2` → `(1,2,1,2)` |

### Tuple Unpacking
```python
x, y, z = (1, 2, 3)
first, *rest = (1, 2, 3, 4)    # first=1, rest=[2,3,4]
*init, last = (1, 2, 3, 4)     # init=[1,2,3], last=4
```

## Dictionaries
Key-value pairs, ordered (Python 3.7+), mutable, keys must be unique.
```python
d = {"name": "Alice", "age": 30}
d = dict(name="Alice", age=30)
d = dict([("name", "Alice"), ("age", 30)])
```

### Accessing & Modifying
```python
d["name"]               # access by key (raises error if missing)
d.get("name")           # access safely (returns None if missing)
d.get("name", "N/A")    # access with default fallback
d["city"] = "NYC"       # add or update key
del d["age"]            # delete a key
```

### Common Dictionary Methods
| Method | Description | Example |
|---|---|---|
| `d.keys()` | All keys | `{"a":1}.keys()` → `dict_keys(["a"])` |
| `d.values()` | All values | `{"a":1}.values()` → `dict_values([1])` |
| `d.items()` | All key-value pairs | `{"a":1}.items()` → `[("a",1)]` |
| `d.get(k, default)` | Safe access | `d.get("x", 0)` → `0` |
| `d.update(d2)` | Merge another dict | `d.update({"b":2})` |
| `d.pop(k)` | Remove & return value | `d.pop("a")` → `1` |
| `d.setdefault(k, v)` | Set if key missing | `d.setdefault("x", 0)` |
| `d.copy()` | Shallow copy | |
| `d.clear()` | Remove all pairs | |

### Iterating
```python
for key in d:
    print(key)

for value in d.values():
    print(value)

for key, value in d.items():
    print(key, value)
```

### Dictionary Comprehensions
```python
squares = {x: x**2 for x in range(5)}
filtered = {k: v for k, v in d.items() if v > 10}
```

### Merging Dicts (Python 3.9+)
```python
merged = d1 | d2          # new merged dict
d1 |= d2                  # merge d2 into d1 in place
```
