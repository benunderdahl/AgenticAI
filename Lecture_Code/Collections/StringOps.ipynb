# String Operations

```python
string = "Hello World"
```

```python
# char

# strings can be indexed with following syntax string[3]
print(string[4]) #returns o

# since it can be indexed we can also loop through every char in the string
for char in string:
    print(char) # will print each char on a new line for the length of the string

# strings are immutable just like in java
```

```python
print(string.upper()) # all upper
print(string.lower()) # all lower
print(string.title()) #capitalizes first letter of each word

print(string.strip()) #removes leading and traling whitespace
# strip can also take arguments for specific chars your trying to remove

# we can split strings as well , helpful for splitting strings
print(string.split()) # this will return two strings in a list


# reassigning a split string lets you do operations on each word that was split
# by default split will use a blank space as teh delimiter, this can be changed 
spl = string.split()

print(spl[1] + spl[0]) # returns WorldHello
```

# Escape Characters

```python
print("This is a random string in double quotes")

# special string encoding
print('\u265e') # returns the corresponding utf-8 special char

```

# String Formatting

```python
x = 103.32
y = 2334.23553

# formatting within print functino start the string with f
# variables can be plugged in with {} brackets

print(f"x is value {x} and y is value {y}") # returns string with values declared and initialized above

# we can round decimal points using :.{floating decimals}f

print(f"x is value {x:.1f} and y is value {y:.2f}") #shotens x to 1 demcial and y to 2 decimals

```

```python
print("dog in dogma") # in is a keyword so will check if value before exists in value after in 

# in is useful for parsing and looking for patterns

# since a string is essentially an array of char we can replace values
# the replace function is useful for this

print("Replace World".replace("Replace", "Hello"))
# Were specifying a value we want to look for, if it exists replace it with the second argument
# Since strings are immutable if you want to use the value that was replaced
# reassign it to another variable

```
