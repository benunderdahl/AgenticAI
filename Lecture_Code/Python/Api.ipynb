## Defining JSON

### stands for Javascript object notation, similar to a python dictionary
### relies on key value pairs and most of the time have nested values
### most common data type for web apps, not normalized

```python
'''
    Strings in JSON are enclosed in double quotes, strict unlike python
'''

employees = [
    {
        "name": 'John',
        "company": {
            "name": 'SkillStorm',
            "location": 'Jacksonville, Fl'
        },
    },
    {
        'Name' : "John",
        'Company': "SkillStorm"
    }
] # example has two object in python array and dictionaries, is of type dict
# is not a JSON object, looks valid but is not
type(employees[0])
```

```python
# we can change the list with dicts above to json with the json module
import json
# will handle converting single quotes to double quotes to follow json convention
out_file = open('json.txt', 'w')

''' 
    Reading and writing JSON data
        methods to write data
        .dump() - converts input into a json-like file
        .dumps() - converts the input into a json string

        methods to read data
        .load() - read in JSON data from a file
        .loads() - read in JSON data from a string
'''
json.dump(employees, out_file, indent=4) # writing into the file we created to json, converted single quotes to double

out_file.close()
```

```python
# using the dumps method

txt = json.dumps(employees)
print(txt) # same as before but just in string format
# can confirm its a string by checking the type
print(type(txt)) # of class str
```

```python
# reading in a json file

with open('json.txt') as f:
    data = json.load(f)

print(data)

data_str = json.loads(txt) # reading in from the string file made earlier
print(data_str) # gives the same output
```

```python
# working with our file system using io module
import io

f = open('json.txt') 
print(type(f)) # class io.textiowrapper
f = io.StringIO('text data') 
print(type(f)) # class io.stringio

f.close()
```

```python
# API requests
# endpoint - url of the api being called
# method is important to specify what type of endpoint were hitting, can have params 
'''
    GET- get info
    POST- store info
    PUT - update info
    DELETE - remove data

    Header - holds metadata for calls 
    Body - holds the raw data being request of sent

    Authentication
        - Api Keys , passed in through the header of the call
        - will have a hashed password that is issued by the owner of the API

        can also access apis using OAuth tokens, more secure
        this token is used to validate access to the api, typically tied to an account
'''
```

```python
# making api requests using the nytimes api
# using the requests module to hit the api endpoints
import requests as re
import os


base_url = "https://api.nytimes.com"
book_api = "/svc/books/v3"
endpoint = "/lists/current/hardcover-fiction.json"
get_req = base_url + book_api + endpoint
print(get_req) # full uri for our request

# stored previously as environ vars
key = os.environ['api-key']
secret = os.environ['api-secret']

# to get a response we have to be authenticated using this key, secret pair
params = {
    'api-key': key,
          }

# saving our request in a var
response = re.get(get_req, params=params) #returns with status code 200
print(type(response)) # response object

# turning our response into json data
data = response.json()
print(data)

# can find the keys to drill down for info we need using keys method
print(data.keys())
# 'status', 'copyright', 'num_results', 'last_modified', 'results'
print(data['status']) # status code of request
print(data['num_results']) # 15 results
print(data['results']) # viewing our data for this case, contains more key value pairs

print(data['results']['books']) # drilling into json getting more specific info for our request

# indexing through this info 
print(data['results']['books'][0]) # returns first object in that array

print(data['results']['books'][0]['title']) # drilling deeper
```

```python
# list comprehension technique for getting all titles for previous call
titles= [x['title'] for x in data['results']['books']]
print(titles) 
```
