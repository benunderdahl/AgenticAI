```python
# using magic methods and classes 

# overriding can be done easily in python but overloading is not
# can inherit fromthe int class and create our own interpretations using the magic methods
# note that python does not have access modifiers like java 
# the convention is _variable name for protected fields or methods and __variablename for private
# python calls double under (dunder) methods implicitly such as the __init__ constructor
class CustomInt(int):
    def __add__(self, other):
        # calling our super class then the __add__ method and making our own implementation
        return super().__add__(other) + 5
    
a = CustomInt(1)
b = CustomInt(2)
print(a + b) # since these are of the type of our custom class they will behave as such

```

```python
# creating a simple class
class MyClass:
    # we can set our fields just like in java and instantiate using a constructor
    # in python the syntax is to use the __init__ method

    def __init__(self):
        self.field = None
    

    # we can also define our own methods using def keyword
    # we have to use the self keyword to say we are using the current object
    # similar to this in java but must be explicitly typed as a param
    def setField(self, field):
        self.field = field

# now i can create an instance of this class and change the fields
subclass = MyClass()
subclass.setField("Using a setter to change field")
print(subclass.field)
```

```python
# using kwargs (keyword arguments) inside our methods

class Person:
    totalPersons = 0 # used in @classmethod
    # making constructor with kwargs
    # we can assign this to a field and it will hold all the kwargs in a dictionary structure
    def __init__(self, name, **kwargs):
        self.name = name
        self.extra = kwargs
        Person.totalPersons += 1 # for every instantiation this will increase, we can find total number of instantiations
    
    # changing methods that are implicit in objects
    # changing the equality method to now compare names of objects for equality
    # offers the flexibility to do comparisons with objects
    def __eq__(self, name):
        if self.name == name:
            return True
        else: 
            return False
        
    # __str__ is called implicitly when inside the print method
    # we can change the output we get for instance if we call an object in the print method
    def __str__(self):
        return f'This object name is {self.name} with kwargs of {self.extra}'
    # creating some custom methods 
    # these methods can have self if we need to reference fields for this instance or not if we dont need it
    def sayHello(self):
        return f'Hello my name is {self.name}'
    
    @classmethod
    def getPersons(cls): # were not looking at an individual objet but rather the class as a whole keyword is cls
        return cls.totalPersons
    
    # magic method for when objects are deleted __del__
    @classmethod
    def __del__(self):
        del self
        Person.totalPersons -= 1

john = Person("John", age=50, city="New York", hobby="Drawing")
# if we call this instance of MyClass1 we can view the kwargs by looking into the extra field
print(john.extra) # useful for taking an arbitrary number of arguments
```

```python
# checking equality from objects instantiated from the same reference
jane = Person("Jane", age=40, city="Milwaukee", hobby="Gardening")
print(john == jane) # this will print False the name is differnet method was changed to have a different
# behavior from the original equality method, overriden 

# if we change janes name to john we will get true
jane.name = "John"
print(john == jane) # returns True

jane.name = "Jane"

# utilizing the changed __str__ method we did above we can get custom output in the print function
print(john) # both list name and kwargs
print(jane)

```

```python

# calling our custom method we created earlier in the Person class
print(john.sayHello())
print(jane.sayHello())
```

```python
'''
Using decorators

@classmethod - pass a reference to the class

@staticmethod - ties a method to the class, some nuance from java static keyword

decorators allow us to modify the behavior of our methods without modifying code inside
decorators are invoked above the methods we want to change the behavior of

'''

# class method can be called from the class level
print(Person.getPersons()) # two instantitions were made above so 2 will be output

# now we can call our delete method and see if it updates the class variable
del jane
print(Person.getPersons()) # output is one jane object was deleted
```
