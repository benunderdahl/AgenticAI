```python
import os
from openai import OpenAI, AzureOpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
# creating params for the model in this case gpt-4o-mini
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content" : "You are a helpful assistant."}, # promp for the system
        {"role": "user", "content": "What color is the sky"}, # prompt or input, also this is our context window the messages array
    ],
    temperature=0.0, # lower temp is more deterministic, and temp of 1 will allow a lot of randomness , range(0-1)
    # turning the temp up essentially results in a broader word bank for the model to choose from
    max_tokens=500, # max token for our response that we get back from the model
)
```

```python
# will return our response from the model api
response.choices

# we can uses following to get stats about our call
response.usage
```

```python
# if we know the model pricing we can get our cost incursion
# by getting the length of tokens and doing simple math
token_count = len(tiktoken.encoding_for_model("gpt-4o-mini").encode("some prompt that will be encoded for a certain model"))
input_cost = token_count / 1_000_000 * PRICING["gpt-4o"]["input"]
output_cost = token_count / 1_000_000 * PRICING["gpt-4o"]["output"]
total = input_cost + output_cost
```
